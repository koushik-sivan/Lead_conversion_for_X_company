```python
X_train_scaled
```

```python
#checking Multicollinearity
#Calculate VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calculate_vif(X_train_scaled):
  vif_data = pd.DataFrame()
  vif_data["Feature"] = X_train_scaled.columns
  vif_data["VIF"] = [variance_inflation_factor(X_train_scaled.values,i) for i in range(len(X_train_scaled.columns))]

  return vif_data
```

```python
#statsmodel
import statsmodels.api as sm  # Importing the statsmodels library

# Add a constant term to the feature matrix for the intercept
X_train_sm = sm.add_constant(X_train_scaled)

# Fit the logistic regression model using statsmodels
sm_model = sm.Logit(y_train, X_train_sm).fit()
```
